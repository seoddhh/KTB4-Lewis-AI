<aside>

```python

    '이름': ['김철수', '이영희', '박민수', '최지현', '홍길동', '정지훈', '이지은'],
```

````python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    '이름': ['김철수', '이영희', '박민수', '최지현', '홍길동', '정지훈', '이지은'],
    '연봉': [4000, 3500, 5000, 4200, 10000, 4600, 3300],
    '가입일': pd.to_datetime(['2020-05-21', '2019-07-15', '2021-01-10', '2018-11-03', '2017-09-27', '2016-04-11', '2022-03-19'])
})

df['수정연봉'] = np.where(df['가입일'].dt.year < 2019, df['연봉'] * 1.1, df['연봉'])

print(df)
print(f"인상전 전체 평균 연봉: {df['연봉'].mean():.2f}")
print(f"전체 평균 연봉: {df['수정연봉'].mean():.2f}")
````

````python
import re


text = "안녕하세요!!! 저는 AI 모델-입니다. 12345 데이터를   정리해 보겠습니다."

cleaned_text = re.sub(r'[^가-힣\s]', '', text)

final_text = re.sub(r'\s+', ' ', cleaned_text).strip()

print(f"원본: {text}")
print(f"정제: {final_text}")
````

### 3번 퀘스트

<aside>

**주어진 텍스트 데이터를 문장 단위로 분리한 후, 각 문장의 단어 개수를 데이터프레임으로 변환하는 코드**를 작성하세요.

```python
# 샘플 텍스트 데이터
text = "자연어 처리는 재미있다. 파이썬과 pandas를 활용하면 편리하다. 데이터 분석은 흥미롭다."
```

````python
import pandas as pd


text = "자연어 처리는 재미있다. 파이썬과 pandas를 활용하면 편리하다. 데이터 분석은 흥미롭다."

sentences = [s.strip() for s in text.split('.') if s.strip()]

word_counts = [len(s.split()) for s in sentences]

df_sentences = pd.DataFrame({'문장': sentences,'단어개수': word_counts})

print(df_sentences)
````
